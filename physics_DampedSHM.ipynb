# Damped Simple Harmonic Motion

simulating damped simple harmonic motion using euler method

```python
import matplotlib.pyplot as plt

# initial condition
init_y=0.3
h=1e-3
y = [init_y]
v=[0]
k = 0.8
t = [0]
m=5
g=9.8

# initial condition
Cd=0.2
A=8
dens=1.2
```

```python
for i in range(500000):

  Fd = -0.5*Cd*A*dens*(v[-1]*abs(v[-1]))

  F = -1*k*y[-1]+Fd
  
  new_y = y[-1] + v[-1]*h + (F*h*h)/(m*2)
  new_v = v[-1] + (F*h)/m
  y.append(new_y)
  v.append(new_v)
  t.append(t[-1]+h)
```

```python
plt.plot(t,y)
plt.title('position-time graph')
plt.show()
```
